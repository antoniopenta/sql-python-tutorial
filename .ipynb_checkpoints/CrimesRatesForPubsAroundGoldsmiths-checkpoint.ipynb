```python
import glob
import pandas as pd
import datetime
from sqlalchemy import create_engine
```

```python
def dateparse(time_in_month):
    time_in_month=str(time_in_month+'-01')
    return datetime.datetime.strptime(time_in_month, "%Y-%m-%d")
```

```python
print('Example of the results when we parse the date string')
print(dateparse('2018-04'))
```

* Crime data downloaded from https://data.police.uk/
* Pub in london downloaded from https://www.getthedata.com/open-pubs/london


```python
l_files = glob.glob('data/crime/*.csv')
df=pd.read_csv(l_files[0],parse_dates=['Month'],date_parser=dateparse)
for file in l_files[1:]:
    df=pd.read_csv(file,parse_dates=['Month'],date_parser=dateparse)
    df.append(df)
df=df.rename(lambda x: '_'.join([ item.lower() for item in x.split(' ')]), axis='columns')
```

```python
def normalize_lsa_name(x):
    if pd.isnull(x):
        return ''
    l=[]
    for item in x.split(' ')[:-1]:
        l.append(item.lower())
    return '_'.join(l)

df['lsoa_name']=df['lsoa_name'].apply(normalize_lsa_name)
print('Data Frame')
df.head(5)
```

```python
print('Number of rows : ',df.shape[0])
# load the data with 
df.to_csv('data/analysis/crime_data_to_upload.csv')
```

--  Query used to create the views, The  position of the GoldSmiths University has been derived by decoding in lat,lng its postode:

~~~~sql 
create view crimes_around_goldsmiths as (select * from crimes as c where 
 6371 * 2 * asin(sqrt(
            POWER(SIN((c.latitude - 51.47414) * pi()/180 / 2),
            2) + cos(c.latitude * pi()/180 ) * COS(51.47414 *
            pi()/180) * power(sin((c.longitude +0.03540) *
            pi()/180 / 2), 2) ))<=3);


create view pubs_around_goldsmiths as (select * from pubs as p where  6371 * 2 * asin(sqrt(
            POWER(SIN((p.latitude - 51.47414) * pi()/180 / 2),
            2) + cos(p.latitude * pi()/180 ) * COS(51.47414 *
            pi()/180) * power(sin((p.longitude +0.03540) *
            pi()/180 / 2), 2) ))<=3);
~~~~

--  Query aggreate each pub with the closest crimes:

~~~~sql 
select p.name_pub,p.latitude,p.longitude, count(*) as total_crime from crimes_around_goldsmiths as c, pubs_around_goldsmiths as p where 
6371 * 2 * asin(sqrt(
            POWER(SIN((p.latitude - c.latitude) * pi()/180 / 2),
            2) + cos(p.latitude * pi()/180 ) * COS(c.latitude *
            pi()/180) * power(sin((p.longitude -c.longitude) *
            pi()/180 / 2), 2) ))<0.1 group by (p.name_pub,p.latitude,p.longitude) having (count(*)>2) order by total_crime desc;
~~~~
