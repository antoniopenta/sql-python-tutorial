```python
import json
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
# downloaded from https://www.thecocktaildb.com/api.php
with open('data/cocktails/cocktails.json') as f:
    data = json.load(f)
pprint(data)
```

```python
l_ingredients = [cocktail[key] for cocktail in data for key in cocktail  if 'Ingredient' in key and cocktail[key]!='']
```

```python
dict_ingrdients = {item: index for index,item in enumerate(list(set(l_ingredients)))}
l_ingredients = [(index,item) for index,item in enumerate(list(set(l_ingredients)))]
df_ingredients=pd.DataFrame.from_records(l_ingredients,columns=['idingredient','stringridient'])
df_ingredients.head(5)
```

```python
cocktail_basic=[(cocktail['idDrink'],cocktail['strDrink'],cocktail['strAlcoholic'],cocktail['strCategory'],cocktail['strGlass'],cocktail['strInstructions']) for cocktail in  data]
```

```python
df_cocktail=pd.DataFrame.from_records(cocktail_basic,columns=['iddrink','strdrink','stralcoholic','strcategory','strglass','strinstructions'])
```

```python
df_cocktail.head(10)
```

```python
l_ingredients_cocktail = []
for cocktail in data:
    #pprint(cocktail)
    for i in range(1,15,1):
        s_ingridients = 'strIngredient'+str(i)
        if cocktail[s_ingridients]!='' and cocktail[s_ingridients] is not None:
            s_portion = 'strMeasure'+str(i)
            assert s_portion!='', 'Portion of an ingredient cannot be null'
            #print(s_ingridients)
            id_ingredients = dict_ingrdients[cocktail[s_ingridients]]
            if  cocktail[s_portion] is None:
                cocktail[s_portion]=''
            l_ingredients_cocktail.append((cocktail['idDrink'],id_ingredients,cocktail[s_portion]))    
```

```python
df_cocktail_ingredients=pd.DataFrame.from_records(l_ingredients_cocktail,columns=['iddrink','idingredient','strmeasure'])
df_cocktail_ingredients['strmeasure'] = df_cocktail_ingredients['strmeasure'].apply(lambda x: x.strip()).replace('', '1')
```

--  Database Schema
~~~~sql

create table cocktail(
iddrink	varchar(10),
strdrink varchar(100),	
stralcoholic varchar(100),
strcategory varchar(100),	
strglass varchar(100),
strinstructions  varchar(10000)  
);
create table ingredients(
    idingredient varchar(100),	
    stringridient varchar(100)

);

create table composition(
id varchar(10),
iddrink	varchar(10),
idingredient varchar(10),	
strmeasure varchar(50)
);


alter table ingredients add primary key(idingredient);
alter table  cocktail add primary key(iddrink);
alter table composition add primary key(id);
alter table composition ADD CONSTRAINT my_fk_1 foreign key (iddrink) references cocktail ON DELETE CASCADE;
alter table composition ADD CONSTRAINT my_fk_2 foreign key (idingredient) references ingredients ON DELETE CASCADE;

~~~~
